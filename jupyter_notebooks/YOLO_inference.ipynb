```python
from ultralytics import YOLO
import matplotlib.pyplot as plt
import cv2
```

```python
img = cv2.imread('/home/nikitina.alina8/WB/trash/hand/pointing-gesture-female-hand-shows-index-finger_77190-3681.jpg')
h, w = img.shape[:2]

min_edge = min(h, w)
start_x = w//2 - min_edge//2
start_y = h//2 - min_edge//2

#   
crop_img = img[start_y:start_y + min_edge, start_x:start_x + min_edge]

#    224224
resized_img = cv2.resize(crop_img, (224, 224))
resized_img_rgb = cv2.cvtColor(resized_img, cv2.COLOR_BGR2RGB)

plt.imshow(resized_img_rgb)
plt.axis('off')
plt.show()
```

```python
cv2.imwrite("test.jpg", resized_img)
```

```python
model = YOLO('./yolo11x/train/weights/best.pt')
```

```python
prediction = model.predict(source=resized_img, save=True)
```
